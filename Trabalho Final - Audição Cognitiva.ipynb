# Avaliação

A Avaliação se dá através da comparação entre o texto original e o texto transcrito pelo algoritmo de Speech to Text.  
A comparação não é feita somente calculando o erro. Para determinar a qualidade (ou a porcentagem de erro) consideramos outros pontos, que são:  
- Palavras substituidas (S);
- Palavras excluidas (D);
- Palavras adicionadas (I);

O cálculo é conhecido como _WER_, ou _Word Error Rate_, e é feito através da seguinte fórmula:  

$$WER = 100 * \frac{S + D + I}{|W|}$$

A Biblioteca **jiwer**, para linguagem Python, realiza o cálculo do _WER_.

Por padrão, a biblioteca realiza alguns pré-processamento que auxiliam para uma medição de erro mais acurada. Os pré-processamentos padrão são:  
- Remoção de espaços em branco;
- Remoção de pontos e virgulas;
- Palavras são colocadas em caixa-baixa;
- Palavras são tokenizadas, ou seja, são separadas individualmente;

```python
# Import
from jiwer import wer
```

### Áudio 1

```python
erro_1 = wer(texto_1, s2t_1)

print('A taxa de error para o Audio 1 é de {0:.2f}%'.format(erro_1*100))
```

### Áudio 2

```python
erro_2 = wer(texto_2, s2t_2)

print('A taxa de error para o Audio 2 é de {0:.2f}%'.format(erro_2*100))
```
